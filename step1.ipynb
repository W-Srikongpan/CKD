##Time stamp

```python
def replace_df(year):
    year = year.replace('2555','2012')
    year = year.replace('2556','2013')
    year = year.replace('2557','2014')
    year = year.replace('2558','2015')
    year = year.replace('2559','2016')
    year = year.replace('2560','2017')
    year = year.replace('2561','2018')
    year = year.replace('2562','2019')
    year = year.replace('2563','2020')
    year = year.replace('2564','2021')
    year = year.replace('2565','2022')
    year = year.replace('2566','2023')
    return year

df_lab['CWhen'] = df_lab['CWhen'].apply(replace_df)
df_lab.head()
```

```python
df_lab['CWhen'] = pd.to_datetime(df_lab['CWhen'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')
df_lab['CWhen']
```

#Tag

```python
CountUnique = '/content/drive/My Drive/ProjectCKD/CountUniqueคนไข้ที่ไม่ฟอกไต072723.xlsx'
df_CountUnique = pd.read_excel(CountUnique ,sheet_name='Lab')
df_CountUnique.head()
```

```python
Laboratory = '/content/drive/My Drive/ProjectCKD/AI for Chronic kidney disease.xlsx'
df_Laboratory = pd.read_excel(Laboratory ,sheet_name='Laboratory')
df_Laboratory.head()
```

    drop column df_Laboratory

```python
delete_Laboratory = ['Lab index no','จน.แถว','Unnamed: 10','Unnamed: 11','Unnamed: 12',
                     'Unnamed: 13','Unnamed: 14','Unnamed: 15','Unnamed: 16','Unnamed: 17',
                     'Unnamed: 18']
df_Laboratory = df_Laboratory.drop(columns = delete_Laboratory, axis=1)
df_Laboratory = df_Laboratory.drop(32)
df_Laboratory = df_Laboratory.drop(33)
df_Laboratory = df_Laboratory.drop(34)
df_Laboratory['Unnamed: 4'] = df_Laboratory['Unnamed: 4'].astype(int)
df_Laboratory['Unnamed: 5'] = df_Laboratory['Unnamed: 5'].astype('Int64')
df_Laboratory['Unnamed: 6'] = df_Laboratory['Unnamed: 6'].astype('Int64')
df_Laboratory['Unnamed: 7'] = df_Laboratory['Unnamed: 7'].astype('Int64')
df_Laboratory['Unnamed: 8'] = df_Laboratory['Unnamed: 8'].astype('Int64')
df_Laboratory['Unnamed: 9'] = df_Laboratory['Unnamed: 9'].astype('Int64')
df_Laboratory
```

```python
df_Laboratory.to_csv('/content/drive/My Drive/ProjectCKD/df_Laboratory.csv', index=False)
```

```python
mapped_values = []
for index, row in df_CountUnique.iterrows() :
    id1 = row['Unnamed: 0']
    matching_row = df_Laboratory[(df_Laboratory['Unnamed: 4'] == id1)| (df_Laboratory['Unnamed: 5'] == id1) |
                                 (df_Laboratory['Unnamed: 6'] == id1)| (df_Laboratory['Unnamed: 7'] == id1) |
                                 (df_Laboratory['Unnamed: 8'] == id1)| (df_Laboratory['Unnamed: 9'] == id1)]
    if not matching_row.empty :
        mapped_values.append(matching_row[True].values[0])
    else:
        mapped_values.append(None)

df_CountUnique['Lab'] = mapped_values
df_CountUnique.dropna(subset=['Lab'], inplace=True)
df_CountUnique
```

    Mapping in df_lab

```python
map_lab = df_CountUnique.set_index('ResultItemName')['Lab'].to_dict()
df_lab['Lab'] = df_lab['ResultItemName'].map(map_lab)
```
