#Censor

    group_eGFR4 is df_eGFR more than and equared 15

```python
group_eGFR4 = group_eGFR1[(group_eGFR1.ResultValue >= 15)]
```

```python
group_eGFR4
```

```python
map_CWhen = df_stage5.set_index('PatientUID')['CWhenfirst'].to_dict()
map_ResultValue = df_stage5.set_index('PatientUID')['ResultValue'].to_dict()
group_eGFR4['CWhe5'] = group_eGFR4['PatientUID'].map(map_CWhen)
group_eGFR4['ResultValue5'] = group_eGFR4['PatientUID'].map(map_ResultValue)
```

```python
group_eGFR4.head(10)
```

    Censored1
        censored = 1
        failure = 0
    Censored2
        censored right = 1
        lost follow-up = 0



```python
def cenored(ResultValue5):
    return 0 if not pd.isna(ResultValue5) and pd.notna(pd.to_numeric(ResultValue5, errors='coerce')) else 1

group_eGFR4['Censored1'] = group_eGFR4['ResultValue5'].apply(cenored)
```

```python
group_eGFR4.head(10)
```

```python
group_eGFR4['Censored2'] = (group_eGFR4.Censored1 == 1) & (group_eGFR2.CWhenfirst >= df_stage5['CWhenfirst'].max())
group_eGFR4['Censored2'] = group_eGFR4['Censored2'].astype(int)
group_eGFR4.head(10)
```
