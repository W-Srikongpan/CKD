#Pivort Table


    Drop df_lab2 when columns Lab is NaN

```python
df_lab2 = df_lab2.dropna(subset=['Lab'])
df_lab2
```

    Change ResultValue is float and word is 0

```python
df_lab2['ResultValue'] = pd.to_numeric(df_lab2['ResultValue'],errors='coerce').fillna(0).astype(float)
```

    Create Pivot table
        index is PatientUID and PatientVisitUID
        columns is Lab
        values is ResultValue
        aggfunc is median (sum,mean,count,min,max,median,std,var)
        fill_value is 0

```python
df_pivot = df_lab2.pivot_table(index=['PatientUID','PatientVisitUID'],columns = 'Lab', values = 'ResultValue',aggfunc = 'last')
df_pivot = df_pivot[list_lab]
df_pivot
```

#Left join with df_eGFR1 and df_pivot by 'PatientUID' and 'PatientVisitUID'

```python
df_eGFR = df_eGFR.merge(df_pivot,on = ['PatientUID','PatientVisitUID'], how = 'left')
df_eGFR.head(10)
```
